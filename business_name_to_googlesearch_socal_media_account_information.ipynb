```python
# Install required libraries
!pip install googlesearch-python python-Levenshtein pandas openpyxl

# Import required libraries
from googlesearch import search
import re
import pandas as pd
from Levenshtein import ratio

# List of allowed domain extensions and social media sites
allowed_extensions = [".com", ".co.uk", ".org", ".digital", ".in"]
social_media_sites = {
    "facebook.com": "Facebook",
    "instagram.com": "Instagram",
    "linkedin.com": "LinkedIn",
    "x.com": "X.com",
    "twitter.com": "Twitter"
}

# Function to extract domain name and extension from a URL
def extract_domain_and_extension(url):
    match = re.search(r"https?://(www\.)?([^/]+)", url)
    if match:
        domain = match.group(2)
        extension = "." + domain.split('.')[-1]
        return domain, extension
    return "", ""

# Function to check if the URL is a social media profile link
def get_social_media_platform(url):
    for site, platform in social_media_sites.items():
        if site in url and url.count('/') <= 4:  # Ensure it's a profile link
            return platform
    return None

# Function to search for links and filter results
def search_links_for_business(business_name):
    query = business_name
    cleaned_business_name = query.lower().replace(" ", "")
    results = {"Business Name": business_name, "Facebook": "", "Instagram": "", "LinkedIn": "", "X.com": "", "Twitter": ""}

    for result in search(query, num_results=20):
        domain, extension = extract_domain_and_extension(result)
        cleaned_domain = domain.split('.')[0].replace("www.", "").lower()
        similarity = ratio(cleaned_business_name, cleaned_domain)

        if similarity >= 0.7 and extension in allowed_extensions and result.count('/') == 3:
            results['Website'] = result
        elif (platform := get_social_media_platform(result)):
            results[platform] = result

    return results

# Read the Excel file containing business names
input_file = "business_names.xlsx"  # Change this to your file path
df = pd.read_excel(input_file)

# Print column names for debugging
print("Columns in Excel file:", df.columns)

# Use the correct column name
column_name = df.columns[0]  # Automatically take the first column
output_data = []

# Loop through each business name and get links
for business in df[column_name]:
    links = search_links_for_business(business)
    output_data.append(links)

# Create a new DataFrame from output data
output_df = pd.DataFrame(output_data)

# Display the result in Google Colab
print(output_df)

# Save the results to a new Excel file
output_file = "business_links_results.xlsx"
output_df.to_excel(output_file, index=False)

print(f"Results saved to {output_file}")
```
